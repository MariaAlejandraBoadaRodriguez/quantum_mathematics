```python
import qiskit
import qiskit_aer
print(qiskit_aer.__version__)
print(qiskit.__version__)

# Importar los componentes necesarios
from qiskit import QuantumCircuit, transpile  # Importar transpile
from qiskit_aer import AerSimulator  # Utilizar AerSimulator directamente
from qiskit.visualization import plot_histogram

# Crear un circuito cuántico con un qubit y un bit clásico
qc = QuantumCircuit(1, 1)

# Aplicar la compuerta Hadamard (H) al qubit
qc.h(0)

# Medir el qubit
qc.measure(0, 0)

# Usar el simulador de Aer para realizar las mediciones
simulator = AerSimulator()

# Transpilar el circuito para el backend específico (simulador)
qc_transpiled = transpile(qc, simulator)  # Utilizar transpile desde qiskit, no desde el simulador

# Ejecutar el circuito en el simulador directamente
result = simulator.run(qc_transpiled, shots=1000).result()

# Obtener el conteo de cuántas veces se obtuvo |0> y cuántas veces se obtuvo |1>
counts = result.get_counts(qc_transpiled)

# Mostrar los resultados en consola
print("Resultado del experimento (1000 mediciones):")
print(counts)

# Mostrar un histograma de los resultados
plot_histogram(counts)
```
